# `spacetime.Worldline` Tutorial

```python
import spacetime as st
import matplotlib.pyplot as plt
import numpy as np
```

`spacetime.Worldline` allows us to create and manipulate worldlines of particles that change velocity over time, as long as they never exceed the speed of light.

We can map the worldline into a different reference frame that is moving at some constant velocity with `spacetime.Worldline.boost`.

Let's create a wordline in a 1+1 spacetime that follows $ x = 0.2 sin(4t) $. We'll generate 300 vertices between the times $ t = 0 $ and $ t = 10 $.

```python
t = np.linspace(0, 10, 300)
x = 0.2 * np.sin(4*t)
events = np.stack([t, x], axis=-1)
w0 = st.Worldline(events)
```

Let's plot the worldline we've created.

```python

# TODO: This is going to be a common operation, so
# probably should add a method like `Worldline.plot()`
w0_plt_data = np.flip(w0._vertices.transpose(), 0)

fig, ax = plt.subplots(figsize=(5,5))

ax.plot(*w0_plt_data)
ax.set_xlim(-6, 6)
ax.set_ylim(0, 11)
plt.xlabel('x-axis')
plt.ylabel('t-axis')
```

This is a space-time diagram of a particle oscillating back and forth.

Let's see how this worldline gets warped if we observe it in a reference frame that is moving at velocity 0.3 times light speed in the positive x-axis direction with respect to the original reference frame. We'll boost the whole worldline by `0.3`.

```python
w1 = w0.boost(0.3)
```

Now let's plot this new worldline on top of the original one to see what changed.

```python

# TODO: This is going to be a common operation, so
# probably should add a method like `Worldline.plot()`
w0_plt_data = np.flip(w0._vertices.transpose(), 0)
w1_plt_data = np.flip(w1._vertices.transpose(), 0)

fig, ax = plt.subplots(figsize=(5,5))

# TODO: This concatenation can be done in a new method `Frame.plot()`
ax.plot(*(np.concatenate([w0_plt_data, w1_plt_data], axis=0)))


ax.set_xlim(-6, 6)
ax.set_ylim(0, 11)
plt.xlabel('x-axis')
plt.ylabel('t-axis')
```
